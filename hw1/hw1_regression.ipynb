# **Extract Features (1)**
![圖片說明](https://drive.google.com/uc?id=1LyaqD4ojX07oe5oDzPO99l9ts5NRyArH)
![圖片說明](https://drive.google.com/uc?id=1ZroBarcnlsr85gibeqEF-MtY13xJTG47)

將原始 4320 * 18 的資料依照每個月分重組成 12 個 18 (features) * 480 (hours) 的資料。 

```python
raw_data.shape
```

```python
month_data={}
for month in range(12):
    sample = np.empty([18, 24*20]) # 18個features 每個月取 20 天, 每天有24小時
    for day in range(20): # 每個月取 20 天
        sample[:, day * 24 : (day + 1) * 24] = raw_data[18 * (20 * month + day) : 18 * (20 * month + day + 1), :] # 以18個特徵為單位存
    month_data[month] = sample
```

# **Extract Features (2)**
![alt text](https://drive.google.com/uc?id=1wKoPuaRHoX682LMiBgIoOP4PDyNKsJLK)
![alt text](https://drive.google.com/uc?id=1FRWWiXQ-Qh0i9tyx0LiugHYF_xDdkhLN)

每個月會有 480hrs，每 9 小時形成一個 data，每個月會有 471 個 data，故總資料數為 471 * 12 筆，而每筆 data 有 9 * 18 的 features (一小時 18 個 features * 9 小時)。

對應的 target 則有 471 * 12 個(第 10 個小時的 PM2.5)

```python
x = np.empty([12 * 471, 18 * 9], dtype = float)
y = np.empty([12 * 471, 1], dtype = float)
```

```python
for month in range(12):
    for day in range(20):
        # print("day:", day)
        for hour in range(24):
            if day == 19 and hour > 14: continue
            # print("hour:", hour)            
            x[month * 471 + day * 24 + hour, :] = month_data[month][:,day * 24 + hour : day * 24 + hour + 9].reshape(1, -1) #vector dim:18*9 (9 9 9 9 9 9 9 9 9 9 9 9 9 9 9 9 9 9)
            y[month * 471 + day * 24 + hour, 0] = month_data[month][9, day * 24 + hour + 9] #value
            
```

# **Normalize (1)**

```python
from sklearn.preprocessing import StandardScaler

# API
scaler = StandardScaler().fit(x)
norm_x = scaler.transform(x)

# custom
mean_x = np.mean(x, axis = 0) #18 * 9 
std_x = np.std(x, axis = 0) #18 * 9 
for i in range(len(x)): #12 * 471
    for j in range(len(x[0])): #18 * 9 
        if std_x[j] != 0:
            x[i][j] = (x[i][j] - mean_x[j]) / std_x[j]
```

# **Training**

(和上圖不同處: 下面的 code 採用 Root Mean Square Error)

因為常數項的存在，所以 dimension (dim) 需要多加一欄；eps 項是避免 adagrad 的分母為 0 而加的極小數值。

每一個 dimension (dim) 會對應到各自的 gradient, weight (w)，透過一次次的 iteration (iter_time) 學習。

```python
learning_rate = 100
eps = 0.0000000001
epoch = 10000
dim = 18 * 9 + 1  # input feature = 18 * 9 prev-hours + 1 bias
```

```python
w = np.zeros([dim, 1])
w.shape
```

```python
x = np.concatenate((np.ones([12 * 471, 1]), x), axis = 1).astype(float) # bias (5652, 1) + x (5652, 162)
adagrad = np.zeros([dim, 1]) # store prev grad
x.shape
```

```python
x.transpose().shape
```

```python
(np.dot(x.transpose(), np.dot(x, w) - y)).shape
```

```python
for t in range(epoch):
    loss = np.sqrt(np.sum(np.power(np.dot(x, w) - y, 2))/471/12) #rmse # [(y' - y)^2 / ( x.shape[0] )] ^ 0.5
    if(t%1000==0):
        print(str(t) + ":" + str(loss))
    gradient = 2 * np.dot(x.transpose(), np.dot(x, w) - y)  # grad = 2 * dot( (y' -y), X_T )   #dim*1   # (163, 1)
    adagrad += gradient ** 2                                # ada_pregrad += gra**2
    w = w - learning_rate * gradient / np.sqrt(adagrad + eps)   # w -= lr * grad / sqrt(ada_pregrad) => remove square
np.save('weight.npy', w)
```

載入 test data，並且以相似於訓練資料預先處理和特徵萃取的方式處理，使 test data 形成 240 個維度為 18 * 9 + 1 的資料。

```python
testdata = pd.read_csv('./test.csv', header = None, encoding = 'big5')
test_data = testdata.iloc[:, 2:]
test_data[test_data == 'NR'] = 0
testdata.head()
```

```python
test_data = test_data.to_numpy()
test_data.shape
```

```python
test_x = np.empty([240, 18*9], dtype = float) # 240 days, 18 features
for i in range(240):
    test_x[i, :] = test_data[18 * i: 18* (i + 1), :].reshape(1, -1)
```

```python
# API (previous scalar)
norm_test_x = scaler.transform(test_x)
print(norm_test_x)

print("--------------------------------------------------------------")
# custom
for i in range(len(test_x)):
    for j in range(len(test_x[0])):
        if std_x[j] != 0:
            test_x[i][j] = (test_x[i][j] - mean_x[j]) / std_x[j]
print(test_x)
```

```python
test_x = np.concatenate((np.ones([240, 1]), test_x), axis = 1).astype(float) # concate bias
test_x.shape
```
